# 1-D Transient Heat Equation Solver
# $\frac{\partial T}{\partial t}=\frac{\kappa}{c_{v}} \frac{\partial^{2} T}{\partial z^{2}}$

For the solution we use a Finite Difference Scheme and more specifically an implicit time discretization is used where the spatial derivatives are written using a central difference scheme at time n+1 and a backward time difference approximation at time n+1.
The computational grid is as follows:
<img src="Compute_Grid_heat.png" style="width:400px;"/>
The finite difference approximation of the $T$ is as follows:
$$
\begin{aligned}
-& \frac{\kappa_{i-1 / 2}}{\Delta z_{i-1 / 2}} T_{i-1}^{n+1}+\left(\frac{c_{v, i} \Delta z_{i}}{\Delta t}+\frac{\kappa_{i-1 / 2}}{\Delta z_{i-1 / 2}}+\frac{\kappa_{i+1 / 2}}{\Delta z_{i+1 / 2}}\right) T_{i}^{n+1} -\frac{\kappa_{i+1 / 2}}{\Delta z_{i+1 / 2}} T_{i+1}^{n+1}=\frac{c_{v, i} \Delta z_{i}}{\Delta t} T_{i}^{n}
\end{aligned}
$$

Boundary Conditions:
we specify $T_0^{n+1}$ at the top     (Dirichlet Boundary)

Zero Flux at bottom $F_{N}^{n+1}=0$ (Neumann Boundary)

Numerical Method used: Implicit Method.
The general form of the tridiagonal system of equations is:

$$
a_{i} T_{i-1}^{n+1}+b_{i} T_{i}^{n+1}+c_{i} T_{i+1}^{n+1}=d_{i}
$$

```julia
global physcon = Soil_Heat_V4.physcon() # Define the number of soil Layers
global soil = Soil_Heat_V4.soil(nsoi=120)
# Generate the Soil Compute Grid
soil = Soil_Heat_V4.compute_grid_settings(soil)

# --- Model run control parameters

tmean = physcon.tfrz + 15.0;       # Mean daily air temperature for diurnal cycle [K]
trange = 10.0;                     # Temperature range for diurnal cycle [K]
dt = 1800;                         # Time step [seconds]
nday = 200;                        # Number of days
soil.soil_texture = 1;             # Soil texture class: sand
soil.method = "excess-heat";       # Use excess heat for phase change

#soil.soil_texture = 11;         # Soil texture class: clay
#soil.method = "apparent-heat-capacity"; # Use apparent heat capacity for phase change
```

```julia
# Initial soil temperature [K] and unfrozen & frozen water [kg H2O/m2]

for i = 1:soil.nsoi

   # Temperature [K]

   soil.tsoi[i] = physcon.tfrz + 2.0

   # Soil water at saturation [kg H2O/m2]

   h2osoi_sat = soil.watsat[soil.soil_texture] * physcon.rhowat * soil.dz[i]

   # Actual water content is some fraction of saturation

   if (soil.tsoi[i] .> physcon.tfrz)
      soil.h2osoi_ice[i] = 0
      soil.h2osoi_liq[i] = 0.8 * h2osoi_sat
   else()
      soil.h2osoi_liq[i] = 0
      soil.h2osoi_ice[i] = 0.8 * h2osoi_sat
   end

end
```

```julia
# --- Time stepping loop to increment soil temperature

# Main loop is NTIM iterations per day with a time step of DT seconds.
# This is repeated NDAY times.

ntim = round(86400/dt)

## This code block is just to store the data for one day
hour_vec = Float64[]
z_vec = Float64[]
tsoi_vec = Float64[]


temp=soil.z
idx=temp.>-1      # First 1m or 100cm
temp1=temp[idx]

hour_out = zeros(Int(ntim))
z_out    = zeros(length(temp1)+1)
tsoi_out = zeros(length(temp1)+1,Int(ntim))

##

for iday = 1:nday
   @printf("day = %6.0f\n",iday)
   for itim = 1:ntim

      # Hour of day

      hour = itim * (dt/86400 * 24)

      # Surface temperature: Constant value TMEAN if TRANGE = 0. Otherwise; use a sine
      # wave with max (TMEAN + 1/2 TRANGE) at 2 pm & min (TMEAN - 1/2 TRANGE) at 2 am

      tsurf = tmean + 0.5 * trange * sin(2*pi/24 * (hour-8.0))

     # Thermal conductivity & heat capacity

      soil = soil_thermal_properties(physcon, soil)
        
        # Organic Matter with Low thermal Conductivity on top 5cm
        #soil.tk[1] = 0.5
        #soil.tk[2] = 0.5

      # Soil temperatures

      soil = soil_temperature(physcon, soil, tsurf, dt)
        
      # Save hourly soil temperature profile for last day

      if (iday == nday)

         # Surface output

         # Vector format - to write to data file
            push!(hour_vec,hour)
            push!(z_vec,0)
            push!(tsoi_vec,tsurf - physcon.tfrz) # deg C

         # For contour()
         hour_out[Int(itim)] = hour
         z_out[1] = 0
         tsoi_out[1,Int(itim)] = tsurf - physcon.tfrz; # deg C

         # Soil layers for top 100 cm

         for i = 1:soil.nsoi
            if (soil.z[i] > -1.0)

                push!(hour_vec,hour)

                push!(z_vec,soil.z[i] * 100) # cm

                push!(tsoi_vec,soil.tsoi[i] - physcon.tfrz)# deg C

               z_out[i+1] = soil.z[i] * 100; # cm
               tsoi_out[i+1,Int(itim)] = soil.tsoi[i] - physcon.tfrz; # deg C
            end
         end

      end

      

   end
end
```

```julia
x = hour_out
y = z_out
X = repeat(reshape(x, 1, :), length(y), 1)
Y = repeat(y, 1, length(x))
PyPlot.contourf(X,Y, tsoi_out) # PyPlot.contourf(X,Y, tsoi_out,cmap="jet")
plt.title("Soil Temperature (deg C)")
plt.xlabel("Time of Day (Hr)")
plt.ylabel("Soil Depth (cm)")
colorbar()
#plt.savefig("line_plotorg.pdf")  
```
